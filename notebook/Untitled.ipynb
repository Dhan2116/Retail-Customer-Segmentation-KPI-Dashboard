```python
import plotly.express as px

def plot_radar(cluster_summary):
    cluster_summary_norm = (
        cluster_summary
        .drop(columns=["NumCustomers"])
        .apply(lambda x: (x - x.min()) / (x.max() - x.min()))
        .reset_index()
        .melt(id_vars="Cluster", var_name="Feature", value_name="Value")
    )

    fig = px.line_polar(
        cluster_summary_norm,
        r="Value",
        theta="Feature",
        color="Cluster",
        line_close=True,
        title="Cluster Profile Comparison"
    )
    fig.show()
```
